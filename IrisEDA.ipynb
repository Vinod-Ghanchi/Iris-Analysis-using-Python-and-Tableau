```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib inline
```

```python
iris_df = pd.read_csv("Iris.csv", index_col='Id')


# Display the first few rows of the dataset
print(iris_df.head())

# Summary statistics
summary_stats = iris_df.describe()
print("\nSummary Statistics:\n", summary_stats)
```

```python
# Visualize Sepal and Petal distributions by Species
plt.figure(figsize=(15, 8))

# Sepal Length distribution
plt.subplot(2, 2, 1)
sns.histplot(data=iris_df, x='SepalLengthCm', kde=True, hue='Species', palette='Set1')
plt.title('Sepal Length Distribution by Species')

# Sepal Width distribution
plt.subplot(2, 2, 2)
sns.histplot(data=iris_df, x='SepalWidthCm', kde=True, hue='Species', palette='Set1')
plt.title('Sepal Width Distribution by Species')

# Petal Length distribution
plt.subplot(2, 2, 3)
sns.histplot(data=iris_df, x='PetalLengthCm', kde=True, hue='Species', palette='Set1')
plt.title('Petal Length Distribution by Species')

# Petal Width distribution
plt.subplot(2, 2, 4)
sns.histplot(data=iris_df, x='PetalWidthCm', kde=True, hue='Species', palette='Set1')
plt.title('Petal Width Distribution by Species')

plt.tight_layout()
plt.show()
```

```python
# Pairwise scatter plot with species color-coding
sns.pairplot(iris_df, hue='Species', palette='Set1')
plt.suptitle('Pairwise Scatter Plot by Species', y=1.02)
plt.show()
```

```python
# Box plots for each feature by species
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_df, palette='Set1')
plt.title('Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=iris_df, palette='Set1')
plt.title('Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=iris_df, palette='Set1')
plt.title('Petal Length by Species')

plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=iris_df, palette='Set1')
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
correlation_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Violin plots for each feature by species
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.violinplot(x='Species', y='SepalLengthCm', data=iris_df, palette='Set1')
plt.title('Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.violinplot(x='Species', y='SepalWidthCm', data=iris_df, palette='Set1')
plt.title('Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.violinplot(x='Species', y='PetalLengthCm', data=iris_df, palette='Set1')
plt.title('Petal Length by Species')

plt.subplot(2, 2, 4)
sns.violinplot(x='Species', y='PetalWidthCm', data=iris_df, palette='Set1')
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
# Swarm plots for each feature by species
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)
sns.swarmplot(x='Species', y='SepalLengthCm', data=iris_df, palette='Set1')
plt.title('Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.swarmplot(x='Species', y='SepalWidthCm', data=iris_df, palette='Set1')
plt.title('Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.swarmplot(x='Species', y='PetalLengthCm', data=iris_df, palette='Set1')
plt.title('Petal Length by Species')

plt.subplot(2, 2, 4)
sns.swarmplot(x='Species', y='PetalWidthCm', data=iris_df, palette='Set1')
plt.title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
# 3D Scatter plot for Sepal dimensions
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}
ax.scatter(iris_df['SepalLengthCm'], iris_df['SepalWidthCm'], iris_df['PetalLengthCm'],
           c=iris_df['Species'].map(colors))

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
plt.title('3D Scatter Plot of Sepal Dimensions')
plt.show()
```

```python
# Parallel Coordinate Plot
from pandas.plotting import parallel_coordinates

plt.figure(figsize=(12, 8))
parallel_coordinates(iris_df, 'Species', colormap='viridis')
plt.title('Parallel Coordinate Plot by Species')
plt.show()
```

```python
# Radar Chart
from math import pi

# Normalize the data
normalized_df = iris_df.groupby('Species').transform(lambda x: (x - x.min()) / (x.max() - x.min()))

# Create a radar chart
categories = list(normalized_df.columns)
values = normalized_df.mean().tolist()
values += values[:1]  # To close the circle
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.fill(angles, values, color='skyblue', alpha=0.7)

ax.set_yticklabels([])
plt.xticks(angles[:-1], categories)
plt.title('Radar Chart of Normalized Feature Averages by Species')
plt.show()
```
